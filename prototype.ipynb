```python
age_data = (finance['age'].value_counts(normalize=True) * 100).reset_index()
age_data.columns = ['age', 'percent']

fig, ax = plt.subplots(figsize=(10,6))
bars = sns.barplot(data=age_data, x='age', y='percent', ax=ax)

ax.set_xlabel("Age")
ax.set_ylabel("Percentage (%)")
ax.set_title("Age Distribution (%)")

ax.bar_label(ax.containers[0], fmt='%.1f%%', fontsize=12)

# -------------------------
# Add two vertical lines
# -------------------------

# Get x positions of ages 22 and 33
x_22 = age_data[age_data['age'] == 29].index[0]
x_33 = age_data[age_data['age'] == 37].index[0]

# Draw vertical lines AT the bars for age 22 and 33
ax.axvline(x=x_22, color='SteelBlue', linestyle='--', linewidth=2)
ax.axvline(x=x_33, color='SteelBlue', linestyle='--', linewidth=2)

plt.tight_layout()
plt.show()
```

```python
gender_data = (finance['gender'].value_counts(normalize=True) * 100).reset_index()
gender_data.columns = ['gender', 'percent']

fig, ax = plt.subplots(figsize=(10,6))

bars = sns.barplot(
    data=gender_data, 
    x='gender', 
    y='percent',
    ax=ax,
)

ax.set_xlabel("Gender")
ax.set_ylabel("Percentage (%)")
ax.set_title("Gender Distribution (%)") 

# Add labels on bars
ax.bar_label(ax.containers[0], fmt='%.1f%%', fontsize = 17)

plt.tight_layout()
plt.show()
```

```python

The gender distribution shows a significant skew: Male investors constitute 62.4%, while Female investors make up only 37.6%. 
This indicates a notable gender gap in participation or representation within the dataset.
Design targeted financial literacy and investment onboarding campaigns for women
```

```python
summary = finance.groupby('investment_avenues')['equity_market'].sum().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=summary, x='investment_avenues', y='equity_market')

ax.set_title("passive ownership")
ax.set_xlabel("INVESTMENT AVENUES(Yes/No)")
ax.set_ylabel("PREFERENCE POINTS in equity market")

# Add value labels
ax.bar_label(ax.containers[0], fmt='%.0f', fontsize = 15)

plt.xticks(fontsize=15)
plt.tight_layout()
plt.show()
```

```python

# Investment avenue columns
cols = ['mutual_funds','equity_market','debentures','government_bonds',
        'fixed_deposits','ppf','gold']

# Calculate averages
avg_scores = finance[cols].mean().reset_index()
avg_scores.columns = ['investment_avenue', 'average_score']

# Plot
plt.figure(figsize=(10,5))
ax = sns.barplot(data=avg_scores, x='investment_avenue', y='average_score')

plt.title('Average Score per Investment Avenue')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

# Add value labels on bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
expect_pct = finance['expect'].value_counts(normalize=True).mul(100).reset_index()
expect_pct.columns = ['expect', 'percentage']

plt.figure(figsize=(10,5))
ax = sns.barplot(data=expect_pct, x='expect', y='percentage')

# Add percentage labels on bars
for p in ax.patches:
    ax.text(
        p.get_x() + p.get_width() / 2,     # x position (center of bar)
        p.get_height() + 0.5,              # y position slightly above bar
        f"{p.get_height():.1f}%",          # label text
        ha='center', fontsize=10
    )

plt.xticks(rotation=45)
plt.title('Return Expectations (%)')
plt.ylabel('Percentage')
plt.show()
```

```python
# Prepare percentage table
savings_pct = (
    finance['what_are_your_savings_objectives?']
    .value_counts(normalize=True)
    .mul(100)
    .reset_index()
)

savings_pct.columns = ['savings_objective', 'percentage']

plt.figure(figsize=(12,5))
ax = sns.barplot(data=savings_pct, x='savings_objective', y='percentage')

# Rotate labels for readability
plt.xticks(rotation=45)
plt.title('Savings Objectives (%)')
plt.ylabel('Percentage')

# Add percentage labels on bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.1f}%',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
